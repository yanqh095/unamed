```python
vali_artery = ['F07', 'F11', 'F17', 'F08', 'F20', 'F19']
import json
import os
import cv2
import numpy as np
spath = '/data1/qhong/seg/mmdetection/data/anno_set/00/test'
tpath = '/data1/qhong/seg/detectron2/arterySeg/artery'
with open('/data1/qhong/seg/mmdetection/data/anno_set/00/annotations/artery_test.json', 'r') as f:
    a = json.load(f)
```

```python
n = len(a['annotations'])
train = {k:[] for k in a.keys()}
train['categories'] = a['categories']
val = {k:[] for k in a.keys()}
val['categories'] = a['categories']
nw = 320
for i in range(n):
    a['images'][i]['width'] = nw
    box = list(a['annotations'][i]['bbox'])
    fn = os.path.join(spath, a['images'][i]['file_name'])
    img = cv2.imread(fn)
    h, w = img.shape[:2]
    w0 = box[2]
    w1 = nw
    c0 = int(np.floor(box[0]))
    c1 = int(np.ceil(c0+w0))
    s = np.random.choice(np.arange(c1-w1-1,c0+1))
    if s+w1 > w:
        s = w - w1 - 10
    if s < 0:
        s = 10
    
    a['annotations'][i]['bbox'][0] -= s
    img = img[:,s:s+w1,:]
    
    if a['images'][i]['file_name'][:3] in vali_artery:
        for k in ['images', 'annotations']:
            val[k].append(a[k][i])
        tfn = os.path.join(os.path.join(tpath, 'val'), a['images'][i]['file_name'])
        cv2.imwrite(tfn, img)
    else:
        for k in ['images', 'annotations']:
            train[k].append(a[k][i])
        tfn = os.path.join(os.path.join(tpath, 'test'), a['images'][i]['file_name'])
        cv2.imwrite(tfn, img)
```

```python
fold = '01'
t = 'test'
old_path = '/data1/qhong/seg/mmdetection/data/anno_set/'
new_path = '/data1/qhong/seg/detectron2/arteryseg/artery/'
fold_path = os.path.join(old_path, fold)
json_file = os.path.join(os.path.join(fold_path, 'annotations'), 'artery_%s.json'%t)
loc = {}

anno_id = 0

with open(json_file, 'r') as f:
    data = json.load(f)
new_data = {}
new_images = []
new_annotations = []
t_path = os.path.join(fold_path, t)
newt_path = os.path.join(os.path.join(new_path, fold), t)
if not os.path.exists(newt_path):
    os.makedirs(newt_path)
for d, anno in zip(data['images'], data['annotations']):
    artery, slc = d['file_name'].split('_')
    artery = int(artery[1:])
    slc = int(slc.split('.')[0])
    image_id = int('1%02d%04d'%(artery, slc))
    new_d = {'id': image_id,
             'artery': artery,
            'file_name': os.path.join(newt_path, d['file_name']),
            'height': 160,
            'width': 320}
    bbox = anno['bbox']
    new_anno = {'bbox': bbox,
               'image_id': image_id,
               'area': bbox[2]*bbox[3],
               'id': anno_id,
               'category_id': 0,
               'lumen': anno['lumen'],
               'outer': anno['outer']}
    anno_id += 1
    new_images.append(new_d)
    new_annotations.append(new_anno)
    if artery in loc:
        loc[artery].append(anno['bbox'][0])
    else:
        loc[artery] = [anno['bbox'][0]]
for artery in loc.keys():
    v = loc[artery]
    vmin, vmax = min(v), max(v)
    loc[artery] = [vmin, vmax]
# annotatons
for image, anno in zip(new_images, new_annotations):
    assert image['id'] == anno['image_id']
    #image 不变
    x_range = loc[image['artery']]
    h, w = 160, 640
    l = np.maximum(0, int(np.ceil(x_range[1]))-2*h+20)
    r = np.minimum(2*h, int(np.floor(x_range[0]))-20)
    s = int((l+r)/2)
    lumen = anno['lumen']
    outer = anno['outer']
    lumen = np.stack([lumen[::2], lumen[1::2]]).transpose()
    outer = np.stack([outer[::2], outer[1::2]]).transpose()
    lumen = polygon2mask((h,w), lumen).astype(np.int32)
    outer = polygon2mask((h,w), outer).astype(np.int32)
    mask = (outer-lumen)[:, s:s+2*h].astype(np.uint8)
    mask = mask_util.encode(np.asarray(mask, order='F'))
    mask['counts'] = mask['counts'].decode('ascii')
    anno['segmentation'] = mask
    anno['bbox'][0] -= s
    del anno['lumen']
    del anno['outer']
anno_path = os.path.join(os.path.join(new_path, fold), 'annotations')
if not os.path.exists(anno_path):
    os.makedirs(anno_path)
cate = [{"id": 0, "name": "artery"}]
new_data = {
    'images': new_images,
    'annotations': new_annotations,
    'categories': cate
}
with open(os.path.join(anno_path,'artery_%s.json'%t), 'w') as f:
    json.dump(new_data, f)
# images
for image in os.listdir(t_path):
    if not image.endswith('jpg'):
        continue
    artery, slc = image.split('_')
    artery = int(artery[1:])
    x_range = loc[artery]
    h, w = 160, 640
    l = np.maximum(0, int(np.ceil(x_range[1]))-2*h+20)
    r = np.minimum(2*h, int(np.floor(x_range[0]))-20)
    s = int((l+r)/2)
    img = imread(os.path.join(t_path, image))[:, s:s+2*h]
    imsave(os.path.join(newt_path, image), img)
```
